```python
numeric_sutunlar = df.select_dtypes(include='number')
df[numeric_sutunlar.columns] = numeric_sutunlar.fillna(numeric_sutunlar.median())
```

```python
import seaborn as sns
sns.boxplot(df["Flavor"])
```

```python
import numpy as np

q1 = df["Flavor"].quantile(0.10)
q3 = df["Flavor"].quantile(0.90)
iqr = q3 - q1

lower_out = q1 - iqr * 1.5
upper_out = q3 + iqr * 1.5

flavor_outliers = (df["Flavor"] < lower_out) | (df["Flavor"] > upper_out)
print(f"Lower outliers -> {lower_out}, Upper outlier -> {upper_out}")
print(df.loc[flavor_outliers, "Flavor"])

# Outlier'lar veri setinden kaldrma
df.drop(df[flavor_outliers].index, inplace=True)
```

```python
from sklearn.metrics import r2_score

X = df[["Aftertaste"]]
y = df["Flavor"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=48)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```

```python
plt.scatter(X_test, y_test, color='blue', label='Gerek Deerler')
plt.scatter(X_test, y_pred, color='red', label='Tahmin Sonular')

plt.xlabel('Aftertaste')
plt.ylabel('Flavor')
plt.title('Tahmin Sonular')
plt.legend()
```
